```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

df=pd.read_csv('경찰청_졸음운전교통사고_2014_2016.csv',encoding='cp949')
df.index
df.shape
df.columns
df.head()
df.info()
# df['구분']=df['구분'].astype('str')
# df.info()
df['년도']=df['구분'].str[:4]
df

df1 =df.groupby(by="년도").sum()
df1.plot(kind = 'bar', rot = 0)
plt.title("년도별 사고횟수")
plt.xlabel("년도")
plt.ylabel("(건)/(명)")
plt.show()

```
