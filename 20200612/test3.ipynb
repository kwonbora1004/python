```python

from wordcloud import WordCloud,STOPWORDS

import numpy as np 
from PIL import Image

import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)


stopwords = set(STOPWORDS)
stopwords.add("")
 
text = open('.txt').read()
little_mask = np.array(Image.open('unnamed.jpg'))

plt.figure(figsize=(8,8))
plt.imshow(little_mask,cmap=plt.cm.gray,interpolation='bilinear')
plt.show()

wc = WordCloud(background_color='white',max_words=1000,mask=little_mask,font_path = "c:/Windows/Fonts/malgun.ttf")
wc = wc.generate(text)


plt.figure(figsize=(8,8))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
