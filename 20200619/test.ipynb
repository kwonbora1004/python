```python
import pandas as pd 
# ,,되있는건 read_csv 
# 파이썬 open
# numpy loadtxt

yob1880 = pd.read_csv('../Data/babynames/yob1880.txt',names=['name','sex','births'])#names 컬럼이름
yob1880

#성별로 그룹화 한다음 , 각 그룸에 대해 birth의 합계 출력 
# yob1880.groupby('sex').births.size() X

yob1880.groupby('sex').births.sum()

#파일 모두 불러 오기 
columns = ['name','sex','births']
pieces = []#리스트안에 프레임을 리스트요소로 저장
for year in range(1880,2011): #데이터 프레임 131
    frame = pd.read_csv('../Data/babynames/yob%d.txt' %year , names = columns)
    frame['year'] = year
    pieces.append(frame)
print(len(pieces))

names = pd.concat(pieces, ignore_index=True)#기존인덱스 무시 

print(names)



#년도 성별에 따라 생일의 합계 index='year',columns='sex'
totalBirths = names.pivot_table('births',index='year',columns='sex',aggfunc=sum)


#시각화 
totalBirths.plot()

#이름이 전체 출생수에서 차지하는 비율을 names에 추가 
# names['porp']=구현
# 연도와 성별로 그룹화 -> 
names
# names['prop'] = 

a = names.groupby(['year','sex']).births.sum()
# a.loc[1880].loc['F']
# a[1880]['F']
# a[0]
a.loc[1880].loc['F']
names.loc[0]['births']/a.loc[1880].loc['F']
round(names.loc[0]['births']/a.loc[1880].loc['F'],3) #소수점 3자리 

def addProp(group): #그룹화된 객체(131 *2) 가 전달됨 
    group['prop'] = group.births/group.births.sum()
    return group

names = names.groupby(['year','sex']).apply(addProp)

names

#names에서 year와 sex컬럼을 기준으로 그룹화한다음 prop 컬럼값에 대한 합계를 출력 
names.groupby(['year','sex']).prop.sum()

#해마다 년도와 성별을 기준으로 그룹했을떄 .. 비중의 최대값
#가장 많이 사용되어진 이름의 비율 
names.groupby(['year','sex']).prop.max()

def getTop1000(group):
    return group.sort_values('births',ascending = False)[:1000]

top1000 = names.groupby(['year','sex']).apply(getTop1000)

top1000.reset_index(drop = True, inplace=True) #drop = True 하면 기존에 있던 sex,year 인덱스도 사라지 inplace = True 하면 변경사항을 top1000에 저장
top1000

#top1000에서 성별이 남자인 데이터를 추출하여 boys
#top1000에서 성별이 남자인 데이터를 추출하여 girls
#boolean 참조 true 인것만 참조 
boys = top1000[top1000['sex'] == 'M']
girls = top1000[top1000['sex'] == 'F']

totalBirths = top1000.pivot_table('births',index = 'year',columns = 'name', aggfunc = sum) 
#생일의 합계 를 구함 인덱스 년도 칼럼 이름  

totalBirths.info()

subset = totalBirths[['John','Harry','Mary','Marilyn']]
subset.plot()

subset.plot(subplots=True,figsize=(12,10)) # 따로 출력 사이즈설정 1 =72픽셀

table = top1000.pivot_table('prop',index ='year' ,columns='sex',aggfunc = sum)
#top1000 상위 1000등이 차지하는 비율 

table.plot()

df = boys[boys.year==2010]

propCumsum = df.sort_values('prop',ascending=False).prop.cumsum() #누적합
type(propCumsum)
propCumsum.values # 배열의 값이 됨
propCumsum.searchsorted(0.5)  
#116 : 0~ 116번까지에 대한 누적합이 0.5를초과 


# 116번쨰 , 115 =0.49xxxxxxxxxxxxxxx 
propCumsum.iloc[116]
#상위 117명의 아이들 이름이 전체에서 50%넘게 차지한다

df = boys[boys.year == 1900]
in1900 = df.sort_values('prop',ascending = False).prop.cumsum()
in1900.searchsorted(0.5) + 1 


names.name
#이름의 마지막 글자의 분포 ? 분포의 변화 ? 
# word = 'Mary'
# word[-1]
# names.name.str[-1]
getLastLetter = lambda x:x[-1]#람다 함수명 주기
lastLetters = names.name.map(getLastLetter)# names.name.map(함수명)
# names.pivot_table('births',index = ,columns=['sex','year'],aggfunc = sum)


lastLetters.name = 'last_letter'
lastLetters
table = names.pivot_table('births',index = lastLetters ,columns=['sex','year'],aggfunc = sum)
# names.pivot_table('births',index = lastLetters ,columns=['sex'],aggfunc = sum)

subtable = table.reindex(columns=[1910,1960,2010],level='year')#인덱스 재지정
subtable









```
