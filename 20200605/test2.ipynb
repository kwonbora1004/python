```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

fig = plt.figure()
x = np.arange(1,5)
a = fig.subplots(2,2)

a[0][0].plot(x, x*x,'red',lw=10)
a[0][0].grid(True) #  
a[0][0].set_title("square")
a[0][1].plot(x,np.sqrt(x))
a[0][1].grid(color='blue',ls='-.',lw=1) #ls , lw 
a[0][1].set_title("square root")
a[1][0].plot(x,np.exp(x))
a[1][1].plot(x,np.log10(x))
plt.show()
```

```python
from pylab import * 
import numpy as np 
import math
import matplotlib.pyplot as plt 

x5 = np.arange(0,math.pi*2,0.05)
y5 = np.sin(x5)
plt.xlim(0,10)
plt.ylim(0,1.5)

plt.xticks([0,2,4,6,8])
plt.yticks([-1,0,1])
plt.plot(x5,y5)
plt.show()

```

```python
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1,0.1,0.8,0.8])
a5.plot(x5,np.exp(x5))
#plt.plot(x5, np.exp(x5))
a5.set_ylabel('exp')
a6 = a5.twinx()
a6.plot(x5,np.log(x5),'rx')
a6.set_ylabel('log')
fig.legend(labels=('exp','log'),loc='upper left')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
data = [[30,25,50,20],[40,23,51,17],[35,22,45,19]]
x6 = np.arange(4)
fig = plt.figure(figsize=(8,4))
ax6 = fig.add_axes([0,0,1,1])

ax6.bar(x6 + 0.00, data[0], color = 'b', width = 0.25)
ax6.bar(x6 + 0.20, data[1], color = 'g', width = 0.25)
ax6.bar(x6 + 0.50, data[2], color = 'r', width = 0.25)
fig.legend(labels=('developer','dba','system engineer'),loc="best")
plt.show()

```

```python
n = 5
menMeans = (20,35,30,35,27)
womenMeans = (25,32,34,20,25)

ind = np.arange(n) #0,1,2,3,4
plt.bar(ind,menMeans,width=0.35, color='r')
plt.bar(ind, womenMeans , width= 0.35 , bottom=menMeans,color = 'b')
plt.xticks(ind,('G1','G2','G3','G4','G5'))
plt.ylabel('Scores')
plt.yticks(np.arange(0,81,10))
plt.legend(labels=['Men','Women'])
plt.show()
```
